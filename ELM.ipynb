```python
import gget

# UniProt accession для человека NRF2
uniprot_id = "Q16236"

# Получаем данные о мотивах из ELM
elm_results = gget.elm(uniprot_id, uniprot=True)

# Смотрим, что вернулось
print(f"Найдено {len(elm_results)} мотивов для {uniprot_id}\n")

# Пример вывода в читаемом виде
for motif in elm_results[:5]:  # покажем первые 5
    print(f"Motif ID: {motif['ELMIdentifier']}")
    print(f"Class: {motif['elm_class']}")
    print(f"Positions: {motif['start']}-{motif['end']}")
    print(f"Regex pattern: {motif['regex']}")
    print(f"Description: {motif['description']}")
    print("-" * 60)
```
